# Matplotlib

```python
def f(x):
    return x**3 - 5*x**2 + 9

import numpy as np
x = np.linspace(-5, 5, num=100)
y = f(x)
import matplotlib.pyplot as plt
plt.plot(x, y)
    
```

# 欧拉常数， 指数函数(Exponential Functions)

```python
def exp(x):
    return np.e**x

print(exp(2))
```

```python
plt.plot(x, exp(x))
```
