```python
from random import uniform
from vector_drawing import draw, Points

u = (-1, 1)
v = (1, 1)

def add(v1, v2):
    return (v1[0]+v2[0], v1[1]+v2[1])

def scale(scale, v1):
    return (scale*v1[0], scale*v1[1])

def random_u():
    return uniform(-3, 3)

def random_s():
    return uniform(-1, 1)

possibilities = [add(scale(random_u(), u), scale(random_s(), v))
                     for i in range(0, 500)]
                 
draw(
    Points(*possibilities)
)
```
